```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Q1-Q6眩晕问题评分的平均值
x=np.array([1,2,3,4,5,6])
y=np.array([2.931034483,2.379310345,3.379310345,3.034482759,2.655172414,2])
fig, ax = plt.subplots()
ax.plot(x,y,'ko--')
# ax.set_xlim()
ax.set_ylim(1,5)
ax.set_xticklabels(['','Q1','Q2','Q3','Q4','Q5','Q6'])
ax.set_xlabel(u'问题编号')
ax.set_ylabel(u'参与者评分')
plt.show()
```

```python
# 加速度眩晕实验
x=np.array([1,2,3,4,5])
y1=np.array([1.761904762,2.19047619,2.476190476,3,3.19047619])
y2=np.array([1.5,2.166666667,2.166666667,2.333333333,2.5])
y3=np.array([1.866666667,2.2,2.6,3.266666667,3.466666667])
fig, ax = plt.subplots()
plt.plot(x,y1,'ko-',label=u'总平均分')
plt.plot(x,y2,'k2-.',label=u'老用户')
plt.plot(x,y3,'ks--',label=u'新用户')
# ax.set_xlim()
ax.set_ylim(1,5)
ax.set_xticks([1,2,3,4,5])
ax.set_xticklabels(['2','4','6','8','10'])
ax.set_xlabel(u'加速度an')
ax.set_ylabel(u'参与者评分')
ax.legend(loc='best')

plt.show()
```

```python
# 平均遗忘率
x=np.array([1,2,3])
y1=np.array([0.078571429,0.059090909,0.05])
y2=np.array([0.045977011,0.021943574,0.017241379])
fig, ax = plt.subplots()
plt.plot(x,y1,'ko-',label=u'实验组')
plt.plot(x,y2,'k2-.',label=u'对照组')
# ax.set_xlim(0,4)
ax.set_ylim(0,0.1)
ax.set_xticks([1,2,3])
ax.set_xticklabels([u'空间对象',u'空间格局',u'空间形状'])
# ax.set_xlabel(u'加速度an')
ax.set_yticklabels([0,2,4,6,8])
ax.set_ylabel(u'参与者平均遗忘率(%)')
ax.legend(loc='best')
plt.show()
```
